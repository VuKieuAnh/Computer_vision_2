```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Loc trung binh
# Lam tron
# Thay the nhung pixcel co do thay doi lon hon so voi trung binh cua mat na
def mean_filter(image, nguong):
    res = np.zeros_like(image)
    for x in range(image.shape[1] - 2):
        for y in range(image.shape[0] - 2):
#             mean = np.mean(image[y: y+3, x: x+3].flatten())
            avg = np.mean(image[y: y+3, x: x+3].flatten())
            if (avg - image[y + 1, x + 1]) > nguong:
              res[y, x]= avg
            else:
              res[y, x] = image[y + 1, x + 1]
    return res
```

```python
path="./anhdemo/trungvi.png"
img = cv.imread(path,0)
new_img= mean_filter(img, 50)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')  
plt.subplot(1, 2, 2)
plt.imshow(new_img, cmap='gray')  
```
